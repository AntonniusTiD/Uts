```python
x = salary [["YearsExperience"]]
y = salary [["Salary"]]

import matplotlib.pyplot as mtp
mtp.scatter(x,y);
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(x,y)

mtp.scatter(x,y)
mtp.plot(x, linreg.predict(x));
```
