```python
import pandas as pd 

from google.colab import drive
drive.mount('/content/gdrive')
directory= '/content/gdrive/MyDrive/survey-lung-cancer.csv'

cv = pd.read_csv(directory)
```

```python
cv.head()
```

```python
cv.info()
```

```python
cv["LUNG_CANCER"].replace(["YES","NO"], [0,1], inplace=True)
cv["GENDER"].replace(["M","F"], [0,1], inplace=True)

x = cv.drop('LUNG_CANCER', axis=1)
y = cv['LUNG_CANCER']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x)
scaled_data = scaler.transform(x)
scaled_data = pd.DataFrame(scaled_data, columns= x.columns)
scaled_data.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=66)
print(len(y_test))
```

```python
from sklearn import linear_model

model = linear_model.LogisticRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
